```python
df.columns=df.columns.str.replace(" ","_")
```

```python
df.columns=df.columns.str.replace('cost_price','Cost_Price')
```

```python
null= df[df['Ship_Mode'].isnull()]
null
```

```python
if df['Ship_Mode'].isna().any():
    df['Ship_Mode']=df['Ship_Mode'].fillna('')
else:
    pass
```
